```
list_of_urls = [
                "http://rigaux.org/language-study/syntax-across-languages/Vrs.html",
                "http://rigaux.org/language-study/syntax-across-languages/Fnctn.html", 
                "http://rigaux.org/language-study/syntax-across-languages/CntrFlow.html",
                "http://rigaux.org/language-study/syntax-across-languages/Tps.html",
                "http://rigaux.org/language-study/syntax-across-languages/bjcrntRfl.html",
                "http://rigaux.org/language-study/syntax-across-languages/PckgMdl.html"
                "http://rigaux.org/language-study/syntax-across-languages/Strng.html",
                "http://rigaux.org/language-study/syntax-across-languages/Blns.html",
                "http://rigaux.org/language-study/syntax-across-languages/BagAndLst.html",
                "http://rigaux.org/language-study/syntax-across-languages/VrsDatTps.html",
                "http://rigaux.org/language-study/syntax-across-languages/Mthmt.html",
                "http://rigaux.org/language-study/syntax-across-languages/Thrds.html"
                ]
```

```
def generateSoup(url):
  page = requests.get(url)
  return BeautifulSoup(page.text, 'html.parser')
```

```
d = {}
table_counter = 1


for url in list_of_urls:
  soup = generateSoup(url)
  table_elements = soup.find_all('table')
  sections = soup.find_all('li')

  for section in sections:
    section_titles = section.findChildren('a', recursive=False)
    # doesnt print numbers in remarks section
    # prob better solution out there
    if (section_titles[0].text[0]) != "(":
      print(section_titles[0].text)

  
  for table in table_elements:
    table_name = "table"+str(table_counter)
    table_counter += 1
    d[table_name] = {}

    table_rows = table.find_all('tr')
    for tr in table_rows:
      td = tr.find_all('td')
      d[table_name][td[0].text] = td[1].text.split(', ')      

```

```
print(d) # can also dump it to json using json.dumps()
```
